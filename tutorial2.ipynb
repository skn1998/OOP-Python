# Class variable and Instance variable

```python
class Student:
    student_count = 0    #class variable
    max_cgpa = -1    #class variable
    
    def __init__(self, name, roll, cgpa):
        self.name = name    # instance variable
        self.roll = roll    # instance variable
        self.cgpa = cgpa    # instance variable
        Student.student_count+=1    # class variable
        Student.max_cgpa = max(Student.max_cgpa, self.cgpa)
```

```python
saurav = Student("Saurav", 49, 8.59)
ravi = Student("Ravi", 40, 9.59)
```

```python
print(saurav)
```

```python
print("Number of students registered is:", Student.student_count)
```

```python
print("Maximum CGPA of student registered is:", Student.max_cgpa)
```

# Order of use of class and object variable

**1. If we are calling an instance variable then the variable will be first searched in the object attributes and if not found then will be searched in the class attributes**

**2. If we are calling a class variable then the variable will be searched in the class attributes directly**

```python
class Employee:
    
    increment = 10000
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def SalaryIncrement(self):
        self.salary += self.increment
    
    def GeneralSalaryIncrement(self):
        self.salary += Employee.increment
```

```python
saurav = Employee("Saurav", 50000)
nishant = Employee("Nishant", 50000)
saurav.increment = 20000
```

```python
saurav.GeneralSalaryIncrement()
nishant.GeneralSalaryIncrement()
```

```python
print(saurav.salary, nishant.salary)
```

```python
saurav.SalaryIncrement()
nishant.SalaryIncrement()
```

```python
print(saurav.salary, nishant.salary)
```
