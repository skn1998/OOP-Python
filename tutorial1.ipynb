# Basics of OOP in python

#### Declaring a class and objects 

```python
class Employee: # declaring a class in python
    pass

# declaring ohjects of class Employee 
saurav = Employee()
nishant = Employee()

# assigning the details to the objects
saurav.name = "Saurav"
saurav.age = 22
saurav.salary = 50000

nishant.name = "Nishant"
nishant.age = 20
```

```python
print(saurav, nishant)
```

#### Extracting all the details of an object

```python
saurav.__dict__ 
```

```python
nishant.__dict__
```

#### Extracting all the details of class

```python
Employee.__dict__
```

#### Accessing the object details using "dot" operator

```python
print("Name:", saurav.name)
print("Age:", saurav.age)
print("Salary:", saurav.salary)
```

```python
print(nishant.name, nishant.age)
```

# Constructor in class

```python
class Student:
    def __init__(self, name, age, cgpa): # constructor (it takes self as a default first argument)
        self.name = name
        self.age = age
        self.cgpa = cgpa
```

```python
ravi = Student("Ravi", 22, 8.59)
ram = Student("Ram", 25, 7.00)
```

```python
print(ravi, ram)
```

```python
print(ravi.name, ravi.age, ravi.cgpa)
```
