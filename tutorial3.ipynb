# Class method and Instance Method

**Class methods are used if we want to change some class variable and Instance methods are used if we want to change some instance varible (But we can also use instance method to change a class variable)**

```python
class Employee:
    
    increment = 10000 # class variable
    
    def __init__(self, name, salary): # constructor
        self.name = name
        self.salary = salary
        
    def SalaryIncrement(self): # instance/object method (it takes an object as a parameter)
        self.salary += Employee.increment
        
    @classmethod # decorator
    def IncrementChange(cls, amount): # class method (it taked a class as a parameter)
        cls.increment = amount
```

```python
saurav = Employee("Saurav", 50000)
```

```python
print("Current Incrememt:", Employee.increment)
saurav.SalaryIncrement()
print("Saurav's salary is:", saurav.salary)
```

```python
Employee.IncrementChange(20000)
print("Current Incrememt:", Employee.increment)
saurav.SalaryIncrement()
print("Saurav's salary is:", saurav.salary)
```
