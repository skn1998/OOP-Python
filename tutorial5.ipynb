# Static method in a class

**1. Class method always take a default argument "cls"**

**2. Instance/Object method always take a default argument "self"**

**3. Static method do not take any default argument and can be called for class as well as for the object**

```python
class Student:
    def _init__(self, name, age):
        self.name = name
        self.age = age
    
    @staticmethod
    def IsOpen(day):
        if day in ["Sunday", "Saturday"]:
            return False
        else:
            return True
```

```python
Student.IsOpen("Sunday")
```

```python
saurav = Student()
saurav.IsOpen("Monday")
```
