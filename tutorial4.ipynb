# Using Class method as an alternative Constructor

```python
class Student:
    def __init__(self, name, age, roll):
        self.name = name
        self.age = age
        self.roll = roll
    
    @classmethod
    def AlternateConstructor(cls, student_string):
        student_string = student_string.split("-")
        return cls(student_string[0], int(student_string[1]), int(student_string[2]))
```

```python
saurav = Student("Saurav", 22, 49)
ravi = Student.AlternateConstructor("Ravi-20-12")
```
