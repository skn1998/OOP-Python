# Setter and Deleter in Python

## Setter

```python
class Student:
    def __init__(self, fname, lname, age):
        self.fname = fname
        self.lname = lname
        self.age = age
        
    @property    
    def email(self):
        return str(self.fname) + "." + str(self.lname) + "@gmail.com"
    
    @email.setter
    def email(self, inp_email):
        fname, lname = inp_email.split("@")[0].split(".")
        self.fname = fname
        self.lname = lname
```

```python
saurav = Student("Saurav", "Nishant", 22)
print(saurav.email)
```

```python
saurav.email = "Nishant.Saurav@gmail.com"
```

```python
print(saurav.__dict__)
```

```python
print(saurav.email)
```

## Deleter

```python
class Student:
    def __init__(self, fname, lname, age):
        self.fname = fname
        self.lname = lname
        self.age = age
        
    @property    
    def email(self):
        if(self.fname == None):
            return "Member not found"
        return str(self.fname) + "." + str(self.lname) + "@gmail.com"
    
    @email.setter
    def email(self, inp_email):
        fname, lname = inp_email.split("@")[0].split(".")
        self.fname = fname
        self.lname = lname
    
    @email.deleter
    def email(self):
        self.fname = None
        self.lname = None
        self.age = None
```

```python
saurav = Student("Saurav", "Nishant", 22)
print(saurav.email)
```

```python
del saurav.email
```

```python
print(saurav.email)
```
