# Inheritence and function overriding

```python
class Employee:
    
    increment = 10000
    
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary
        
    def SalaryIncrement(self):
        self.salary += Employee.increment
```

```python
class Developer(Employee):
    def __init__(self, name, age, salary, lang, exp):
        super(). __init__(name, age, salary) # here we are calling the constructor of the base class
        self.lang = lang
        self.exp = exp
    
    def SalaryIncrement(self):  # this function overrides the function in the base class
        self.salary += (Developer.increment + 5000)
```

```python
saurav = Employee("Saurav", 22, 50000)
ravi = Developer("Ravi", 21, 50000, "Python", "10 years")
```

```python
saurav.SalaryIncrement()
ravi.SalaryIncrement()
```

```python
print("Saurav's salary is:", saurav.salary)
print("Ravi's salary is:", ravi.salary)
```
