# Property Decorator

**Property decorators allow us to use a method also a variable**

```python
class Student:
    def __init__(self, fname, lname, age):
        self.fname = fname
        self.lname = lname
        self.age = age
        
    @property    
    def email(self):
        return str(self.fname.lower()) + str(self.lname.lower()) + "@gmail.com"
```

```python
saurav = Student("Saurav", "Nishant", 22)
```

```python
print(saurav.email)
```

#### Note:
**print(saurav.email()) : this will show error as we are calling email as a function**
